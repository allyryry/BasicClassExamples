```python
#Filename is "fifa_eda.csv"
df = pd.read_csv("fifa_eda.csv")
df.head()
```

```python
#take basic stats min, max etc
df.describe(include="all")
```

```python
#get rid of people making no money
df = df[df["Wage"] > 0]
```

```python
#look at people that make the most money (defaults to ascending so have to do true or false to ascending)
df = df.sort_values(by=["Wage", "Age"], ascending=False)
df.head(10)
```

```python
#how to find one value you are looking for 
df[df["Wage"]== 565]
```

```python
#create graphs AND CHOPPING OUT ONE COLUMN
wages = pd.Series(df["Wage"])
wages.head()
```

```python
#CREATE hist (round to nearest million hence -1)
hist = thinkstats2.Hist(round(wages) -1)
```

```python
#pmf
pmf = thinkstats2.Pmf(round(wages) -1)
```

```python
#cdf
cdf = thinkstats2.Cdf(wages)
```

```python
#plot objects from last cell
thinkplot.PrePlot(3, cols=3)
thinkplot.Hist(hist)
thinkplot.SubPlot(2)
thinkplot.Pmf(pmf)
thinkplot.SubPlot(3)
thinkplot.Cdf(cdf)
thinkplot.Config()
```

```python
# if it is a straight line then it is a pareto, visual check, need to specific column if not chopping it out.
logWage = np.log(wages)
logCDF = thinkstats2.Cdf(logWage)

thinkplot.Cdf(logCDF, scaley="log",complement=True)
```

```python
#check to see what proportion is where it starts to flatten out on pmf (20m)
cdf.Prob(20)
```

```python
#take a look at the people below 20 million, copy paste from previus cell to make graphs
lowWage = df[df["Wage"] < 20]
lowhist = thinkstats2.Hist(lowWage["Wage"])
lowpmf = thinkstats2.Pmf(lowWage["Wage"])
lowcdf = thinkstats2.Cdf(lowWage["Wage"])

thinkplot.PrePlot(3, cols=3)
thinkplot.Hist(lowhist)
thinkplot.SubPlot(2)
thinkplot.Pmf(lowpmf)
thinkplot.SubPlot(3)
thinkplot.Cdf(lowcdf)
thinkplot.Config()
```

```python
#vis checks
loglowWage = np.log(lowWage["Wage"])
loglowCDF = thinkstats2.Cdf(loglowWage)
thinkplot.PrePlot(2, cols=2)
thinkplot.Cdf(logCDF, scaley="log",complement=True)
thinkplot.SubPlot(2)
thinkstats2.NormalProbabilityPlot(loglowWage)
thinkplot.Config()
```
